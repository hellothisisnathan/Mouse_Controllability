Dictionary output (hopefully more readable in CSV):

```python
# find all structure unionizes for every wild_type experiment and get info on the listed "injection_structures" + the whole brain
for experiment in wild_type['id'].tolist():
    # Get list of injection_structures
    injection_structures = wild_type.loc[experiment]['injection_structures']
    # Get all injection volumes and densities etc. for the injection_structures + the whole brain to compare
    # Can also get the volumes in every brain region for each injection, would just take a long time
    structure_unionizes = mcc.get_structure_unionizes([experiment], 
                                                  is_injection=True,
                                                  structure_ids=[*injection_structures, 997],
                                                  include_descendants=False)
    # Find total injection volume using the entire brain entry
    inj_vol = structure_unionizes.loc[(structure_unionizes['hemisphere_id'] == 3) & (structure_unionizes['structure_id'] == 997)]['projection_volume'].iloc[0]
    inj_str_vol = dict.fromkeys(injection_structures, 0)
    # Find each percentage
    for k,v in inj_str_vol.items():
        inj_str_vol[k] = structure_unionizes.loc[(structure_unionizes['hemisphere_id'] == 3) & (structure_unionizes['structure_id'] == k)]['projection_volume'].iloc[0]
    injection_structures_percentages = {k: round(v / inj_vol, 6) for k, v in inj_str_vol.items()}
    
    # Round volumes
    for k, v in inj_str_vol.items():
            inj_str_vol[k] = round(v, 6)
    # Add columns to dataframe
    wild_type.at[experiment, 'injection_structures_volumes'] = inj_str_vol                   
    wild_type.at[experiment, 'injection_volume_percentages'] = injection_structures_percentages
```
