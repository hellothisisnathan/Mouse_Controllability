```python


from allensdk.core.mouse_connectivity_cache import MouseConnectivityCache

# The manifest file is a simple JSON file that keeps track of all of
# the data that has already been downloaded onto the hard drives.
# If you supply a relative path, it is assumed to be relative to your
# current working directory.
mcc = MouseConnectivityCache(manifest_file='connectivity/mouse_connectivity_manifest.json')

# open up a list of all of the experiments
all_experiments = mcc.get_experiments(dataframe=True)
print("%d total experiments" % len(all_experiments))

# take a look at what we know about an experiment with a primary motor injection
all_experiments.loc[181057754]

```

```python
wild_type = all_experiments.loc[(all_experiments['strain'] == 'C57BL/6J') & (all_experiments['transgenic_line'].isnull())]
print('Length = %i' % len(wild_type))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(wild_type)
```

```python
nodes = []
for a, row in wild_type.iterrows():
    if row['structure_id'] not in nodes:
        nodes.append(row['structure_id'])
        
print('# unique nodes in set = %i' % len(nodes))
```

```python
wild_type.to_csv('experiment data.csv')
```

```python
coords = {}
for i, j in wild_type.iterrows():
    coords[i] = len(mcapi.get_experiment_detail(i)[0]['specimen']['stereotaxic_injections'][0]['stereotaxic_injection_coordinates'])
```

```python
coords
```

```python
exps_and_injs = {}
for i, j in wild_type.iterrows():
    exps_and_injs[i] = mcapi.get_experiment_detail(i)[0]['specimen']['stereotaxic_injections'][0]['stereotaxic_injection_coordinates']
```

```python
exps_and_injs
```

## For some reason, the experiements with only one injection have an 'id' different from the experiment id, but the experiments with 2 injections have an id equal to their experiment id!!

```python
#wild_type = all_experiments.loc[(all_experiments['strain'] == 'C57BL/6J') & (all_experiments['transgenic_line'].isnull())]
multi_inj_list = [k for k, v in coords.items() if v > 1]
#multiple_injs = wild_type.loc[wild_type['id'].isin(multi_inj_list)]['id']
multiple_injs = pd.DataFrame()
for exp, injs in exps_and_injs.items():
    if exp in multi_inj_list:
        exp_struct_ab = all_experiments.loc[exp]['structure_abbrev']
        exp_struct_id = all_experiments.loc[exp]['structure_id']
        exp_struct_name = all_experiments.loc[exp]['structure_name']
        for inj in injs:
            inj['structure_abbrev'] = exp_struct_ab
            inj['structure_id'] = exp_struct_id
            inj['structure_name'] = exp_struct_name
            multiple_injs = multiple_injs.append(inj, ignore_index=True)
column_titles = ['id', 'structure_abbrev', 'structure_id', 'structure_name', 'angle', 'coordinates_ap', 'coordinates_dv', 'coordinates_ml', 'registration_point', 'stereotaxic_injection_id']
multiple_injs = multiple_injs[column_titles]
multiple_injs
```

```python
multiple_injs.to_csv('experiments with 2 injections.csv', index=False)
```
