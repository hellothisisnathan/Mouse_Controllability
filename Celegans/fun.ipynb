```python
import networkx as nx
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("NeuronConnect.csv")
df
```

```python
neurons = set().union(df['Neuron 2'].unique().tolist(),df['Neuron 1'].unique().tolist())
len(neurons)
```

```python
G = nx.Graph()
G.add_nodes_from(neurons)
```

```python
for a, b in df[['Neuron 1', 'Neuron 2']].itertuples(index=False):
    G.add_edge(a,b)
G.number_of_nodes(), G.number_of_edges()
```

```python
# https://github.com/chahak13/nxcontrol
# Controllability addon for networkx based on Liu et al.

# https://github.com/danielracz/thesisMScPub
# Hungarian guy's thesis code
```

```python
#nx.draw_spectral(G, with_labels=True, font_weight='bold', figsize=(1200,1200), nodesize=10)
pos = nx.spring_layout(G, k=0.5, iterations=20)
nx.draw(G, pos, figsize=(30, 30), with_labels=True, nodesize=20)
plt.savefig('graph.png', dpi=1000)
plt.show()
```

```python
adj = nx.to_pandas_adjacency(G)  # Adjacency matrix
ed = nx.to_pandas_edgelist(G)  # Edge list
adj
```
